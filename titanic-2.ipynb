```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv('/kaggle/input/titanic/train.csv')
train_df.info()
```

```python
train_df
```

```python
train_df.isna().sum()
```

```python
age_mean = train_df['Age'].mean()
train_df['Age'].fillna(value = age_mean, inplace = True)
```

features not needed : PassengerId, Name, Ticket, Fare, emabarked 
features needed : survived, pclass, sex, age, sibsp, parch, cabin

```python
y_train = train_df["Survived"]
y_train
# xtrain = x
```

```python
x_train = train_df.drop(['Survived', 'PassengerId'], axis = 1)
x_train
```

```python
x_train['Fare'].corr(y_train)
```

```python
x_train['Cabin'].nunique()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
x_train['Cabin'] = le.fit_transform(x_train['Cabin'])
x_train['Cabin']
```

```python
x_train['Cabin'].corr(y_train)
```

```python
x_train
```

```python
temp = pd.get_dummies(x_train, columns = ['Sex'])
x_train = temp
x_train
# Male = True, Female = False
```

```python
temp_2 = pd.get_dummies(x_train, columns = ['Embarked'])
x_train = temp_2
x_train
```

```python
x_train['Parch'].nunique()
```

```python
x_train.drop(['Ticket', 'Name'], inplace = True, axis = 1)
x_train
```

```python
x_train.isna().sum()
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy', max_depth=4)
dtc??
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x_train, y_train, random_state = 1)
X_train
```

```python
Y_train
```

```python
dtc.fit(X_train, Y_train)
```

```python
Y_pred = dtc.predict(X_test)
Y_pred
```

```python
ans_pred = pd.DataFrame(data = Y_pred)
```

```python
ans_pred.to_csv('ans.csv', index = False)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_pred, Y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'criterion' : ('entropy', 'gini', 'log_loss'), 'max_depth' : [1,2,3,4,5,6,7,None]}
grid_clf = GridSearchCV(dtc, parameters)
grid_clf.fit(X_train, Y_train)
```

```python
sorted(grid_clf.cv_results_.keys())
grid_clf.best_estimator_
```

```python
grid_clf.predict(X_test)
```

```python
accuracy_score(grid_clf.predict(X_test), Y_test)
```
